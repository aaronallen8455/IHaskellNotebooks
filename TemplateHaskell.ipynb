# Template Haskell

Here's how to write a function that will create on demand a function that will produce a version of `curry` that works to the n'th degree.

```haskell
:ext TemplateHaskell
:m Control.Monad
:m Language.Haskell.TH

curryN :: Int -> ExpQ
curryN n = do
  f <- newName "f"
  xs <- replicateM n $ newName "x"
  let args = map VarP (f:xs)
      ntup = TupE $ map VarE xs
  return $ LamE args (AppE (VarE f) ntup)
  
let curry2 = $(curryN 2)
let curry3 = $(curryN 3)

f (a, b) c = a + b + c
:t curry2 f

g (a, b, c) d = a + b + c + d
:t curry3 g

:t conT
:t tupleDataName
```
